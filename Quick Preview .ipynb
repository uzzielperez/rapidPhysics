# Discriminating Variables 

* $ \sigma_{i\eta i\eta}   $
* $ R_9 $
* H/E

````python
hSieie = phoHistsFile.Get("h_pho_sigmaIEtaIEta") 
hFSieie = phoHistsFile.Get("hFake_pho_sigmaIEtaIEta")

hR9 = phoHistsFile.Get("h_pho_R9")
hFR9 = phoHistsFile.Get("hFake_pho_R9")

hHoverE = phoHistsFile.Get("h_pho_HoverE")
hFHoverE = phoHistsFile.Get("hFake_pho_HoverE")
````

````python
cSieie = ROOT.TCanvas("canvasSieie")
cSieie.cd()

hSieie.Draw()
hFSieie.SetLineColor(2)
# hFSieie.SetFillColor(2)
hFSieie.Draw("SAME")

legend = ROOT.TLegend(0.7,0.6,0.85,0.75) 
legend.AddEntry(hSieie ,"Prompt") 
legend.AddEntry(hFSieie ,"Fake/Unmatched") 
legend.SetLineWidth(0) 
legend.Draw("same")

cSieie.Draw()
````

````python
cR9 = ROOT.TCanvas("canvasR9")
cR9.cd()

hR9.Draw()
hFR9.SetLineColor(3)
hFR9.SetFillColor(2)
hFR9.Draw("SAME")

legend = ROOT.TLegend(0.7,0.6,0.85,0.75) 
legend.AddEntry(hR9 ,"Prompt") 
legend.AddEntry(hFR9 ,"Fake/Unmatched") 
legend.SetLineWidth(0) 
legend.Draw("same")

cR9.Draw()
````

````python
cHoE = ROOT.TCanvas("canvasHoE")
cHoE.cd()

hHoverE.Draw()
hFHoverE.SetLineColor(3)
hFHoverE.SetFillColor(2)
hFHoverE.Draw("SAME")
hHoverE.GetXaxis().SetRangeUser(0, 0.2);

legend = ROOT.TLegend(0.7,0.6,0.85,0.75) 
legend.AddEntry(hHoverE ,"Prompt") 
legend.AddEntry(hFHoverE ,"Fake/Unmatched") 
legend.SetLineWidth(0) 
legend.Draw("same")

cHoE.Draw()
````
